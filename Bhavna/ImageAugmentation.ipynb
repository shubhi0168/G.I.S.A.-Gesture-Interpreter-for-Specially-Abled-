```python
import cv2
import numpy as np
j=1
z=9
def flip_image(image,direction,j,z):
    image=cv2.flip(image,direction)
    cv2.imwrite("/Newdataset/" +str(z)+ "/" +str(j) +".jpeg" ,image)
def rotate_image(image,deg,j,z):
    rows, cols,c = image.shape
    M = cv2.getRotationMatrix2D((cols/2,rows/2), deg, 1)
    image = cv2.warpAffine(image, M, (cols, rows))
    cv2.imwrite("/Newdataset/"+str(z)+"/"+str(j)+".jpeg", image) 
for i in range(1,205):
    image=cv2.imread("/Dataset/"+str(z)+"/" +str(i) +".JPG")
    cv2.imwrite("/Newdataset/"+str(z)+"/" +str(j) +".jpeg" ,image)
    j=j+1
    darkimg=image[:,:,:]*.75 #darkiamage
    cv2.imwrite("/Newdataset/"+str(z)+"/" +str(j) +".jpeg" ,darkimg)
    j=j+1
    brightimg=image[:,:,:]*1.1 #brightimage
    cv2.imwrite("/Newdataset/"+str(z)+"/" +str(j) +".jpeg" ,brightimg)
    j=j+1
    flip_image(image,1,j,z)
    j=j+1
    rotate_image(image,20,j,z)
    j=j+1
    rotate_image(image,-20,j,z)
    j=j+1
```
